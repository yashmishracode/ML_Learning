```python
########################## All Together #################

# Step-1: 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv("C:\\Users\\yashm\\OneDrive\\Desktop\\nareshIT-DataScience\\data file\\redwine.csv")

# Step-2: 

# Input data
X=df.drop('quality',axis=1)
# Output data
y=df['quality']

# Step-3: 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,
                                                  test_size=0.2,
                                                  random_state=1234  )

# Step-4: 

X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

$Step-5$:

**Develop The ML Model**

- Read the package
- Save the package
- apply the fit transform

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
LR
```

$Step-6$:

**Get the predictions**

will pass X_train to get prediction

```python
y_predictions=LR.predict(X_test)
```

```python
#comparing 5 values of y_test(Actual output) and y_predictions(predicted output)
y_test[:5].values,y_predictions[:5]
```

$Step-07$:
**Metrics:** Evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,y_predictions))
print(mean_squared_error(y_test,y_predictions))
print(r2_score(y_test,y_predictions))
```

- mean Square Error manually
- Understand the formulae
- sum(y_test,y_prediction)^2

- will take each observation
- will do subtraction
- will do the square
- summation
- summantion/n

```python
import numpy as np
summ=0
for i in range(len(y_test)):
    summ = summ+(y_test.values[i]-y_predictions[i])**2

summ/len(y_test)
```
